# Combining Both Regions into One CSV

```python
import pandas as pd
```

```python
bej = pd.read_csv('bejaia_region_clean.csv')
sid = pd.read_csv('sidibelabbes_region_clean.csv')
```

```python
# Adding the region columns
bej['region'] = 'bejaia'
sid['region'] = 'sidibelabbes'
print(bej.head(3))
print(sid.head(3))
```

```python
# Stacking the dataframes on top of each other
combo = pd.concat([bej, sid], axis = 0).reset_index(drop = True)
combo
```

```python
#writing out to csv
combo.to_csv('regional_clean.csv', index = False)
df = pd.read_csv('regional_clean.csv')
df
```
